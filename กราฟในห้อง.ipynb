```python
import matplotlib.pyplot as plt

class Person:
    def __init__(self, name, exercise_hours, weight_loss):
        self.name = name
        self.exercise_hours = exercise_hours  # ชั่วโมงออกกำลังกายต่อสัปดาห์
        self.weight_loss = weight_loss        # น้ำหนักที่ลดได้ (กก.)


p1 = Person("Alice", 2, 0.5)
p2 = Person("Bob", 4, 1.2)
p3 = Person("Charlie", 6, 2.0)
p4 = Person("David", 8, 3.0)
p5 = Person("Eva", 10, 4.5)

people = [p1, p2, p3, p4, p5]


exercise = [p.exercise_hours for p in people]
weight_loss = [p.weight_loss for p in people]


plt.figure(figsize=(8,6))
plt.scatter(exercise, weight_loss, color='blue', marker='o')


plt.title('Exercise Hours vs. Weight Loss')
plt.xlabel('Exercise Hours per Week')
plt.ylabel('Weight Loss (kg)')
plt.grid(True)


plt.show()
```
