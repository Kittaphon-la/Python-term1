```python
import matplotlib.pyplot as plt

class Fruit:
    def __init__(self, name, origin, unit_sold):
        self.name = name
        self.origin = origin
        self.unit_sold = unit_sold

# ข้อมูลผลไม้
fruit1 = Fruit("Apple", "USA", 40)
fruit2 = Fruit("Banana", "Thailand", 70)
fruit3 = Fruit("Mango", "India", 55)
fruit4 = Fruit("Orange", "Spain", 30)
fruit5 = Fruit("Durian", "Thailand", 15)

fruits = (fruit1, fruit2, fruit3, fruit4, fruit5)


labels = [f.name for f in fruits]
sales = [f.unit_sold for f in fruits]


plt.plot(labels, sales, marker='o', color='green')  
plt.xlabel('Fruit')
plt.ylabel('Units Sold')
plt.title('Fruit Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```
